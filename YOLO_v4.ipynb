```python
from flask_ngrok import run_with_ngrok
from pyngrok import ngrok
from flask import Flask, render_template,flash, request
from werkzeug.utils import secure_filename
import os
import cv2
import shutil

app = Flask(__name__)
app.config['SECRET_KEY']='secure'
app.config['UPLOAD_FOLDER']='/content/drive/MyDrive/Mini_project/data/Flask-model/static/upload_images/'
app.config['STATIC_FOLDER']='static/images/'
ngrok.set_auth_token('2BFBxUZzzK4jToiCeq1BImmI7wN_5Bis34WP1F3WcZNJJwsfh')
logo_path=os.path.join(os.path.join(app.config['STATIC_FOLDER']), 'logo.png')
bg_path=os.path.join(os.path.join(app.config['STATIC_FOLDER']), 'bg.jpg')
run_with_ngrok(app)
@app.route("/")
def home():
    change_model()
    return render_template('index.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)

@app.route("/test",methods=["GET","POST"])
def test():
           file= request.files['inputfile']
           file_name = secure_filename(file.filename)
           if file_name=='':
                 flash('No file attached!')
                 return render_template('index.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)
           mimetype = file.content_type
           mimetype=mimetype.split('/')
           if mimetype[0]!='image':
               flash("Unsupported format, pls upload jpg/jpeg/png/tfif file")
               return render_template('index.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)
           file.save(os.path.join(app.config['UPLOAD_FOLDER'], file_name))
           org_img='static/upload_images/'+file_name
           path=os.path.join(os.path.join(app.config['UPLOAD_FOLDER']), file_name)
           change_darknet()
           verify(path)
           shutil.copy('predictions.jpg','/content/drive/MyDrive/Mini_project/data/Flask-model/static/predictions/')
           old_name='/content/drive/MyDrive/Mini_project/data/Flask-model/static/predictions/predictions.jpg'
           new_name='/content/drive/MyDrive/Mini_project/data/Flask-model/static/predictions/'+file_name
           os.rename(old_name,new_name)
           change_model()
           path_predicted='static/predictions/'+file_name
           return render_template('index.html',filename1=org_img,filename2=path_predicted,logo=logo_path,b_g=bg_path)

@app.route("/models")
def models():
  return render_template('models.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)

@app.route("/dataset")
def dataset():
  return render_template('dataset.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)

@app.route("/team")
def team():
  return render_template('team.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)

@app.route("/contact")
def contact():
  return render_template('contact.html',filename2='',filename1='',logo=logo_path,b_g=bg_path)

@app.route("/compare")
def compare():
  path='static/images/compare.png'
  return render_template('compare.html',filename2='',filename1=path,logo=logo_path,b_g=bg_path)

app.run()
```

```python
# define helper functions
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()

# use this to upload files
def upload():
  from google.colab import files
  uploaded = files.upload()
  for name, data in uploaded.items():
    with open(name, 'wb') as f:
      f.write(data)
      print ('saved file', name)

# use this to download a file
def download(path):
  from google.colab import files
  files.download(path)
```
